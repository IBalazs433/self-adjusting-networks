# Binary Search Tree

### Imports

```python
import sys
sys.path.append("..")

from trees.bst import BinarySearchTree
from visualization.graph_visualization import plot_tree
```

### Empty Tree

```python
tree = BinarySearchTree()

assert tree.root is None
assert tree.search(5) is None
```

### Insert Root

```python
tree = BinarySearchTree()

node = tree.insert(10)

assert tree.root is node
assert node.key == 10
assert node.parent is None
```

### Insert Left and Right Children

```python
tree = BinarySearchTree()

root = tree.insert(10)
left = tree.insert(5)
right = tree.insert(15)

assert root.left is left
assert root.right is right

assert left.parent is root
assert right.parent is root

assert left.is_left_child
assert right.is_right_child
```

### BST Structure

```python
tree = BinarySearchTree()

for x in [10, 5, 15, 3, 7, 12, 20]:
    tree.insert(x)

assert tree.root.key == 10

assert tree.root.left.key == 5
assert tree.root.right.key == 15

assert tree.root.left.left.key == 3
assert tree.root.left.right.key == 7

assert tree.root.right.left.key == 12
assert tree.root.right.right.key == 20
```

### Search Existing and Missing Nodes

```python
tree = BinarySearchTree()

for x in [10, 5, 15, 3]:
    tree.insert(x)

assert tree.search(3).key == 3
assert tree.search(7) is None
```

### Search Cost

```python
tree = BinarySearchTree()

for x in [10, 5, 15, 3]:
    tree.insert(x)

assert tree.search_cost == 0

tree.search(3)

assert tree.search_cost == 3
```

### Visualization

```python
tree = BinarySearchTree()

for x in [10, 5, 15, 3, 7]:
    tree.insert(x)

plot_tree(tree, title="BST Example")
```

# Splay Tree

### Imports

```python
from trees.splay_tree import SplayTree
from visualization.graph_visualization import plot_tree
```

### Rotate Right

(Node 30)

```python
tree = SplayTree()

n50 = tree.insert(50)
n30 = tree.insert(30)
n20 = tree.insert(20)
n40 = tree.insert(40)
n35 = tree.insert(35)

plot_tree(tree)

tree.rotate_right(n30)

assert n30.parent is None

assert n30.left == n20
assert n20.parent == n30

assert n30.right == n50
assert n50.parent == n30

assert n50.left == n40
assert n40.parent == n50

assert n40.left == n35
assert n35.parent == n40

plot_tree(tree)
```

### Rotate Left

(Node 50)

```python
tree = SplayTree()

n30 = tree.insert(30)
n50 = tree.insert(50)
n40 = tree.insert(40)
n35 = tree.insert(35)

plot_tree(tree)

tree.rotate_left(n50)

assert n50.parent is None

assert n50.left == n30
assert n30.parent == n50

assert n30.right == n40
assert n40.parent == n30

assert n40.left == n35
assert n35.parent == n40

plot_tree(tree)
```

### Zig

```python
tree = SplayTree()

n20 = tree.insert(20)
n10 = tree.insert(10)

tree.splay(n10)

assert tree.root == n10
assert n10.right == n20
assert n20.parent == n10
```

### Zag

```python
tree = SplayTree()

n10 = tree.insert(10)
n20 = tree.insert(20)

tree.splay(n20)

assert tree.root == n20
assert n20.left == n10
assert n10.parent == n20
```

### ZigZig

```python
tree = SplayTree()

n50 = tree.insert(50)
n30 = tree.insert(30)
n10 = tree.insert(10)

tree.splay(n10)

assert tree.root == n10

assert n10.right == n30
assert n30.parent == n10

assert n30.right == n50
assert n50.parent == n30
```

### ZagZag

```python
tree = SplayTree()

n10 = tree.insert(10)
n30 = tree.insert(30)
n50 = tree.insert(50)

tree.splay(n50)

assert tree.root == n50

assert n50.left == n30
assert n30.parent == n50

assert n30.left == n10
assert n10.parent == n30
```

### ZigZag

```python
tree = SplayTree()

n50 = tree.insert(50)
n10 = tree.insert(10)
n30 = tree.insert(30)

tree.splay(n30)

assert tree.root == n30

assert n30.left == n10
assert n30.right == n50

assert n10.parent == n30
assert n50.parent == n30
```

### ZagZig

```python
tree = SplayTree()

n10 = tree.insert(10)
n50 = tree.insert(50)
n30 = tree.insert(30)

tree.splay(n30)

assert tree.root == n30

assert n30.left == n10
assert n30.right == n50

assert n10.parent == n30
assert n50.parent == n30
```

### Search

```python
tree = SplayTree()

for key in [50,30,70,20,40,60,80,10,65]:
    tree.insert(key)

plot_tree(tree)
```
