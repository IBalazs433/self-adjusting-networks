# Binary Search Tree

### Imports

```python
from bst import BinarySearchTree

import sys
sys.path.append("..")
from visualization.graph_visualization import plot_tree
```

### Empty Tree

```python
tree = BinarySearchTree()

assert tree.root is None
assert tree.search(5) is None
```

### Insert Root

```python
tree = BinarySearchTree()

node = tree.insert(10)

assert tree.root is node
assert node.key == 10
assert node.parent is None
assert node.is_root
```

### Insert Left and Right Children

```python
tree = BinarySearchTree()

root = tree.insert(10)
left = tree.insert(5)
right = tree.insert(15)

assert root.left is left
assert root.right is right

assert left.parent is root
assert right.parent is root

assert left.is_left_child
assert right.is_right_child
```

### BST Structure

```python
tree = BinarySearchTree()

for x in [10, 5, 15, 3, 7, 12, 20]:
    tree.insert(x)

assert tree.root.key == 10

assert tree.root.left.key == 5
assert tree.root.right.key == 15

assert tree.root.left.left.key == 3
assert tree.root.left.right.key == 7

assert tree.root.right.left.key == 12
assert tree.root.right.right.key == 20
```

### Search Existing and Missing Nodes

```python
tree = BinarySearchTree()

for x in [10, 5, 15, 3]:
    tree.insert(x)

assert tree.search(3).key == 3
assert tree.search(7) is None
```

### Search Cost

```python
tree = BinarySearchTree()

for x in [10, 5, 15, 3]:
    tree.insert(x)

assert tree.search_cost == 0

tree.search(3)

assert tree.search_cost == 2
```

### Visualization

```python
tree = BinarySearchTree()

for x in [10, 5, 15, 3, 7]:
    tree.insert(x)

plot_tree(tree, title="BST Example")
```

# Optimal Static Binary Search Tree

### Imports

```python
from optimal_bst import build_optimal_bst

import sys
sys.path.append("..")
from visualization.graph_visualization import plot_tree
```

### Single Key

```python
tree = build_optimal_bst([10], [5])

assert tree.root.key == 10
assert tree.root.left is None
assert tree.root.right is None
```

### Three Keys

```python
tree = build_optimal_bst(
    [10, 20, 30],
    [3, 10, 3]
)

assert tree.root.key == 20
assert tree.root.left.key == 10
assert tree.root.right.key == 30
```

### Larger example

```python
tree = build_optimal_bst(
    [10, 20, 30, 40, 50],
    [34, 8, 50, 21, 16]
)

assert tree.root.key == 30

plot_tree(tree)
```
